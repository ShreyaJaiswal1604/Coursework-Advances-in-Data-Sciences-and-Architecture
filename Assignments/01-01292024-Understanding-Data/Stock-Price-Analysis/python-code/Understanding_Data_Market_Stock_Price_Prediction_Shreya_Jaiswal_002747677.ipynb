   - Many websites and web services provide application programming interfaces (APIs), which allow you to explicitly request data in a structured format. This saves you the trouble of having to scrape them.

4. **JSON (and XML)**
   - Because HTTP is a protocol for transferring text, the data you request through a web API needs to be serialized into a string format.
   - Often this serialization uses JavaScript Object Notation (JSON).
   - JavaScript objects look quite similar to Python dicts.
   - Sometimes an API provider only provides responses in XML.

       - **Statistical Context**: This type of distribution suggests that each value in your data has an equal chance of occurring. For example, if you roll a fair six-sided die, each number (1 through 6) has an equal chance of being rolled, leading to a uniform distribution.
     - **Normal (Gaussian) Histogram**:
       - **Explanation**: Picture a scenario where most values in your data are clustered around a central point, forming a symmetrical bell-shaped curve. In a normal histogram, the middle values have higher frequencies, and as you move away from the center, frequencies decrease.
       - **Statistical Context**: This distribution is commonly found in natural phenomena. For instance, if you measure the heights of a large group of people, you'll likely find that most people have heights close to the average, forming a bell curve. The normal distribution is characterized by its mean (average) and standard deviation (spread).

**FUNCTION 1: standardize_column_names**

---
The function `standardizeColumnNames(df)` standardizes column names in a DataFrame by performing the following steps:

1. **Remove Trailing and Leading Spaces**: Trims leading and trailing spaces from column names using the `strip()` method.
2. **Remove Extra Spaces**: Replaces consecutive spaces in column names with a single space using the `str.replace()` method.
3. **Replace Spaces with Underscores**: Replaces spaces in column names with underscores to ensure consistency.
4. **Lowercase Column Names**: Converts all column names to lowercase using the `lower()` method.
5. **Return Standardized DataFrame**: Returns the DataFrame with standardized column names.

The function helps ensure uniformity and consistency in column names, making it easier to work with the DataFrame across different operations.


```python
def standardizeColumnNames(df):
  #remove trailing and leading spaces
  df=df.rename(columns=lambda x :  x.strip())

  #remove extra spaces
  df.columns = df.columns.str.replace('  ',' ')

  #replace space with underscore
  df.columns=df.columns.str.replace(' ','_')

  #lowercase the column names
  df = df.rename(columns=lambda x: x.lower())

  return df
```

**FUNCTION 2: get_yfinance_date_range**

---
The function `get_yfinance_date_range(years_back=1)` retrieves the start and end dates for querying data using the yfinance library.

- **Input**:
  - `years_back` (int): Number of years back from the current year to determine the start date.

- **Output**:
  - `start_date` (datetime): The start date for data retrieval.
  - `end_date` (datetime): The end date for data retrieval, which is the current date.

The function calculates the start date based on the provided number of years back from the current year and sets the end date to the current date.


```python

def get_yfinance_date_range(years_back=1):
    """
    Get the start and end dates for retrieving data using yfinance.

    Args:
    - years_back (int): Number of years back from the current year.

    Returns:
    - start_date (datetime): The start date for data retrieval.
    - end_date (datetime): The end date for data retrieval (current date).
    """
    # Get the current year
    current_year = datetime.now().year

    # Calculate the start date
    start_date = datetime(current_year - years_back, 1, 1)

    # End date is the current date
    end_date = datetime.now()

    return start_date, end_date
```

**FUNCTION 3: get_stock_data**

---
The function `get_stock_data(symbol, start_date, end_date)` retrieves stock data for a given symbol from Yahoo Finance.

- **Input**:
  - `symbol` (str): Ticker symbol of the company.
  - `start_date` (str or datetime): Start date for data retrieval.
  - `end_date` (str or datetime): End date for data retrieval.

- **Output**:
  - `stock_data` (DataFrame): DataFrame containing stock data for the specified symbol.

The function first fetches data from Yahoo Finance using the `yf.download()` function, specifying the symbol, start date, and end date. Then, it standardizes the column names of the retrieved DataFrame using the `standardizeColumnNames()` function. Finally, it returns the DataFrame containing the stock data.


```python
def get_stock_data(symbol, start_date, end_date):
    """
    Retrieve stock data for a given symbol from Yahoo Finance.

    Args:
    - symbol (str): Ticker symbol of the company.
    - start_date (str or datetime): Start date for data retrieval.
    - end_date (str or datetime): End date for data retrieval.

    Returns:
    - stock_data (DataFrame): DataFrame containing stock data for the specified symbol.
    """
    # Fetch data from Yahoo Finance
    stock_data = yf.download(symbol, start=start_date, end=end_date)
    stock_data = standardizeColumnNames(stock_data)
    return stock_data
```

---

**LSTM MODELING**

Long Short-Term Memory (LSTM) modeling is a type of recurrent neural network (RNN) architecture designed to handle sequential data and overcome the vanishing gradient problem in traditional RNNs. LSTM networks are particularly effective for processing and making predictions based on time series or sequential data.

The key features of LSTM networks include:

1. **Memory Cells**: LSTM networks contain memory cells that can maintain information over time intervals. These memory cells allow the network to remember long-term dependencies in the input sequences.

2. **Gates**: LSTM networks have different types of gates (input gate, forget gate, output gate) that regulate the flow of information into and out of the memory cells. These gates help in controlling the information flow and deciding which information to retain or forget.

3. **Forget Gate**: The forget gate determines which information from the previous time step should be discarded from the memory cell.

4. **Input Gate**: The input gate decides which new information from the current time step should be stored in the memory cell.

5. **Output Gate**: The output gate determines what information from the memory cell should be passed on to the next layer or used as the output of the LSTM unit.

LSTM networks are widely used in various applications, including:

- **Time Series Prediction**: LSTM models are commonly used to forecast future values based on historical time series data, such as stock prices, weather patterns, and sensor data.

- **Natural Language Processing (NLP)**: LSTMs are used for tasks such as language translation, sentiment analysis, text generation, and speech recognition, where the input data is sequential in nature.

- **Speech Recognition**: LSTMs are used in speech recognition systems to analyze and interpret spoken language.

Overall, LSTM modeling is a powerful technique for capturing temporal dependencies and making predictions in sequential data, making it suitable for a wide range of tasks in machine learning and artificial intelligence.

**LSTM EQUATION**

---

$$
\begin{equation}
\boxed{
\begin{aligned}
i_t &= \sigma(W_{xi}x_t + W_{hi}h_{t-1} + b_i) \\
f_t &= \sigma(W_{xf}x_t + W_{hf}h_{t-1} + b_f) \\
o_t &= \sigma(W_{xo}x_t + W_{ho}h_{t-1} + b_o) \\
\tilde{C}_t &= \tanh(W_{xc}x_t + W_{hc}h_{t-1} + b_c) \\
C_t &= f_t \odot C_{t-1} + i_t \odot \tilde{C}_t \\
h_t &= o_t \odot \tanh(C_t)
\end{aligned}
}
\end{equation}
$$



Where:

- \( i_t \), \( f_t \), \( o_t \): Input, forget, and output gates respectively.
- \( \tilde{C}_t \): Candidate cell state.
- \( C_t \): Cell state.
- \( h_t \): Hidden state.
- \( x_t \): Input at time step \( t \).
- \( h_{t-1} \): Hidden state from the previous time step.
- \( W \): Weight matrices.
- \( b \): Bias vectors.
- \( \sigma \): Sigmoid activation function.
- \( \odot \): Element-wise multiplication.
