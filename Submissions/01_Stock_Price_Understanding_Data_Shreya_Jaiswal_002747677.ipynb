       **Statistical Context** - This type of distribution suggests that each value in your data has an equal chance of occurring. For example, if you roll a fair six-sided die, each number (1 through 6) has an equal chance of being rolled, leading to a uniform distribution.

      * **Normal (Gaussian) Histogram**:
      
       **Explanation**: Picture a scenario where most values in your data are clustered around a central point, forming a symmetrical bell-shaped curve. In a normal histogram, the middle values have higher frequencies, and as you move away from the center, frequencies decrease.

       **Statistical Context**: This distribution is commonly found in natural phenomena. For instance, if you measure the heights of a large group of people, you'll likely find that most people have heights close to the average, forming a bell curve. The normal distribution is characterized by its mean (average) and standard deviation (spread).

```python
import pandas as pd
import io
import csv
import requests
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# URL for the stock CSV file
url = 'https://raw.githubusercontent.com/ShreyaJaiswal1604/Coursework-Advances-in-Data-Sciences-and-Architecture/main/Data/01-stock-price/stocks.csv'
```

```python
# Fetch the content of the CSV file from the URL
response = requests.get(url)
content = io.StringIO(response.text)
```

---
> 1. Rename columns
---

FUNCTION: rename columns

```python
def clean_columns(df):
    df.columns = [col.lower().replace(' ', '_') for col in df.columns]
    return df
```

```python
df_clean = clean_columns(df)
print(df_clean.columns)
```

---

2. Remove irrelevant columns

---

```python
# Remove irrelevant columns
irrelevant_columns = ['adj_close']  # Add any other columns you consider irrelevant
df_clean = df_clean.drop(columns=irrelevant_columns)
```

---

3. Check and Convert required data types

---

```python
# Check and convert required data types
df_clean['date'] = pd.to_datetime(df_clean['date'])  # Convert 'Date' to datetime format
df_clean.info()
```

---

4. Check for missing values

---

```python
# Check for missing values
missing_values = df_clean.isnull().sum()
print("Missing Values:")
print(missing_values)
```

---

5. Check for duplicate values

---

```python
# Check for duplicate values
duplicate_rows = df_clean[df_clean.duplicated()]
print("\nDuplicate Rows:")
print(duplicate_rows)
```

---
6. Check for Outliers and handle the same
---

```python
# Check for outliers and handle them (using z-score method as an example)
from scipy.stats import zscore
z_scores = zscore(df_clean[['open', 'high', 'low', 'close', 'volume']])
z_scores
```

```python
df_no_outliers = df_clean[(z_scores < 3).all(axis=1)]  # Keep only rows without outliers
display(df_no_outliers)

df_final = df_no_outliers
```

---
Note:

---
* The Z-score is a statistical measure that describes the number of standard deviations a particular data point is from the mean.
* The threshold of 3 is a common choice; it implies that data points with a Z-score greater than 3 or less than -3 are considered outliers.

IV. **Manipulating Data**

---

---
USE CASE : highest-ever closing price for each symbol

---

```python
from collections import defaultdict

# Assuming df_final is your DataFrame
data = df_final.to_dict(orient='records')
data
```

```python
# Group rows by symbol
by_symbol = defaultdict(list)
for row in data:
    by_symbol[row["symbol"]].append(row)

by_symbol
```

```python
for symbol, grouped_rows in by_symbol.items():
  print(symbol)
  print(grouped_rows)
```

```python
# Use a dict comprehension to find the max closing price for each symbol
max_price_by_symbol = {
    symbol: max(row["close"] for row in grouped_rows)
    for symbol, grouped_rows in by_symbol.items()
}
```

```python
max_price_by_symbol
```

```python
list(max_price_by_symbol.items())
```

```python
# Convert the result to a DataFrame for easier plotting
max_prices_df = pd.DataFrame(list(max_price_by_symbol.items()), columns=['Symbol', 'Max Closing Price'])

# Plot the bar chart
max_prices_df.plot(kind='bar', x='Symbol', y='Max Closing Price', legend=False)
plt.title('Max Closing Price for Each Symbol')
plt.xlabel('Symbol')
plt.ylabel('Max Closing Price')
plt.show()
```
